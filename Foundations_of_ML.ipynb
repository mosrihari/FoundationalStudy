# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(88)
X = np.linspace(0, 10, 50)
true_m , true_c = 2, 1  # my expected slope and intercept
y_true = true_m * X + true_c + np.random.normal(0,2, X.shape)
```

```python
X.shape, y_true.shape
```

```python
plt.scatter(X, y=y_true, c='blue')
```

```python
ITERATIONS = 2000
LEARNING_RATE = 0.02
m = np.random.random(1)
c = np.random.random(1)
```

```python
y_pred = m*X + c
(1/len(X)) * np.sum((y_true-y_pred)**2)
```

```python
(-2/len(X)) * np.sum(X * (y_true - y_pred))
```

```python
y_pred = m*X + c
plt.scatter(X, y=y_true, c='blue')
plt.scatter(X, y=y_pred, c='red')
```

```python
for i in range(ITERATIONS):
  y_pred = m*X + c
  MSE = (1/len(X)) * np.sum((y_true-y_pred)**2)
  dMSE_dm = (-2/len(X)) * np.sum(X * (y_true - y_pred))
  dMSE_dc = (-2/len(X)) * np.sum(y_true - y_pred)

  m = m - LEARNING_RATE * dMSE_dm
  c = c - LEARNING_RATE * dMSE_dc

  if i % 100 == 0: # print every 100 iterations
    print(f"ITERATION:{i}, Loss: {MSE} with m={m} and c={c}")
```

```python
y_pred = m*X + c
plt.scatter(X, y=y_true, c='blue')
plt.scatter(X, y=y_pred, c='red')
```

# Gradient Descent using Matrix Multiplication method
This is efficient coding method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100, 1) * 10
y_true = 3 * X + 7 + np.random.normal(0,2, X.shape)
```

```python
theta = np.random.randn(2,1)
```

```python
theta # m and c
```

```python
y_pred = theta[0][0]*X + theta[1][0] # initial plot
plt.scatter(X, y=y_true, c='blue')
plt.scatter(X, y=y_pred, c='red')
```

```python
np.concatenate([X, np.ones((X.shape[0],1))], axis=1)[:3], X[:3], np.concatenate([X, np.ones((X.shape[0],1))], axis=1).shape
```

```python
ITERATIONS = 1000
LEARNING_RATE = 0.02
loss_history = []
X = np.concatenate([X, np.ones((X.shape[0],1))], axis=1) # add a second column for X
for i in range(ITERATIONS):
  y_pred = np.matmul(X,theta) # 100*2 matmul with 2*1 and the result is 100*1
  errors = y_true - y_pred # 100*1 result error
  mse_loss = np.mean(np.sum(errors**2))
  loss_history.append(mse_loss)
  gradient = (-2/len(X)) * np.matmul(X.T, errors) # X.T is 2*100 and errors is 100*1 and the results are 2 gradients m and c each

  theta -= LEARNING_RATE * gradient #scaler * (2*1 matrix) result in 2*1 for theta (shape of theta remains the same but the values are updated)

  if i % 100 == 0: # print every 100 iterations
    print(f"ITERATION:{i}, Loss: {mse_loss} with m={theta[0][0]} and c={theta[1][0]}")
```

```python
plt.scatter(X[:,0], y=y_true, c='blue')
plt.scatter(X[:,0], y=y_pred, c='red') # new y_pred
```

# Stochastic Gradient descent
Where we choose a random point and error for gradient update instead of all points

```python
X = np.random.rand(100, 1) * 10
y_true = 3 * X + 7 + np.random.normal(0,2, X.shape)
theta = np.random.randn(2,1)
```

```python
theta # m and c
```

```python
y_pred = theta[0][0]*X + theta[1][0] # initial plot
plt.scatter(X, y=y_true, c='blue')
plt.scatter(X, y=y_pred, c='red')
```

```python
ITERATIONS = 1000
LEARNING_RATE = 0.02
loss_history = []
X = np.concatenate([X, np.ones((X.shape[0],1))], axis=1) # add a second column for X
for i in range(ITERATIONS):
  y_pred = np.matmul(X,theta) # 100*2 matmul with 2*1 and the result is 100*1
  errors = y_true - y_pred # 100*1 result error
  mse_loss = np.mean(np.sum(errors**2))
  loss_history.append(mse_loss)

  # pick a sample random point
  random_point = np.random.randint(0, X.shape[0])
  random_x = X[random_point].reshape(X.shape[1],-1) #shape is 2*1
  random_error = errors[random_point] # shape is (1,)
  gradient = (-2/len(X)) * np.matmul(random_x, random_error).reshape(X.shape[1], -1) # random_x is 2*1 and errors is (1,) and the result will be (2,) and hence reshape to (2,1) to match theta shape

  theta -= LEARNING_RATE * gradient #scaler * (2*1 matrix) result in 2*1 for theta (shape of theta remains the same but the values are updated)

  if i % 100 == 0: # print every 100 iterations
    print(f"ITERATION:{i}, Loss: {mse_loss} with m={theta[0][0]} and c={theta[1][0]}")
```

```python
plt.scatter(X[:,0], y=y_true, c='blue')
plt.scatter(X[:,0], y=y_pred, c='red') # new y_pred
```

```python
plt.plot(loss_history)
```

# Gradient Descent with momentum
Instead of relying on gradient for the loss, here the velocity is considered for theta update
theta = theta - learning_rate * velocity

where velocity(new) = beta * velocity(old) + (1-beta) * gradient

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic_loss_function(x, y):
  return (x**2) + 10 * (y**2)
```

```python
def grad_cal(x, y):
  dl_dx = 2*x
  dl_dy = 20*y
  return np.array([dl_dx, dl_dy])
```

```python
def gradient_descent_batch(gradient_function, x, y, epochs, learning_rate):
  starting_point = x, y
  path = [(x,y)]
  loss_history = []

  for i in range(epochs):
    loss = quadratic_loss_function(x,y)
    gradient = gradient_function(x,y)
    x -= learning_rate * gradient[0]
    y -= learning_rate * gradient[1]
    loss_history.append(loss)
    path.append((x,y))

  return loss_history, np.array(path)
```

```python
starting_point = (1.5, 1.5)
epochs = 150
learning_rate = 0.1
loss_history, path = gradient_descent_batch(grad_cal, starting_point[0], starting_point[1],
                       epochs, learning_rate)
```

```python
plt.plot(loss_history)
```

```python
def gradient_descent_with_momentum(gradient_function, x, y, beta, epochs, learning_rate):
  starting_point = x, y
  vt = np.array([0,0])
  path = [(x,y)]
  loss_history = []

  for i in range(epochs):
    loss = quadratic_loss_function(x,y)
    gradient = gradient_function(x,y)
    vt = beta * vt + (1-beta) * gradient
    x -= learning_rate * vt[0]
    y -= learning_rate * vt[1]
    loss_history.append(loss)
    path.append((x,y))

  return loss_history, np.array(path)
```

# RMSProp
Where the learning rate is adaptive

```python
def gradient_descent_with_rmsprop(gradient_function, x, y, beta, epochs, learning_rate, epsilon):
  starting_point = x, y
  Egt2 = np.array([0,0])
  path = [(x,y)]
  loss_history = []

  for i in range(epochs):
    loss = quadratic_loss_function(x,y)
    gradient = gradient_function(x,y)
    Egt2 = beta * Egt2 + (1-beta)*(gradient**2)
    multiplication_value_0 = learning_rate / (np.sqrt(Egt2[0]) + epsilon)
    multiplication_value_1 = learning_rate / (np.sqrt(Egt2[1]) + epsilon)
    x -= multiplication_value_0 * gradient[0]
    y -= multiplication_value_1 * gradient[1]
    if(i < 10):
      print(f"{i} : x --> {x} and y --> {y}")
    loss_history.append(loss)
    path.append((x,y))

  return loss_history, np.array(path)
```

# Adam
Combination of RMSProp and Momentum

```python
def gradient_descent_with_adam(gradient_function, x, y, beta1, beta2, epochs, learning_rate, epsilon):
  starting_point = x, y
  Egt2 = np.array([0,0])
  mt = np.array([0,0])
  path = [(x,y)]
  loss_history = []

  for t in range(1, epochs+1):
    loss = quadratic_loss_function(x,y)
    gradient = gradient_function(x,y)

    # update Egt2 value
    Egt2 = beta2 * Egt2 + (1-beta2)*(gradient**2)
    # scale Egt2 value
    Egt2_scaled = Egt2 / (1 - (beta2**t))

    # update mt value
    mt = beta1 * mt + (1-beta1)*(gradient)
    # scale mt value
    mt_scaled = mt / (1 - (beta1**t))

    # update theta
    multiplication_value_0 = learning_rate / (np.sqrt(Egt2_scaled[0]) + epsilon)
    multiplication_value_1 = learning_rate / (np.sqrt(Egt2_scaled[1]) + epsilon)
    x -= multiplication_value_0 * mt_scaled[0]
    y -= multiplication_value_1 * mt_scaled[1]

    if(t < 10):
      print(f"{t} : x --> {x} and y --> {y}, m --> {mt_scaled[0], mt[0], mt[1], mt_scaled[1]}, Egt2 --> {Egt2_scaled[0], Egt2_scaled[1]}")
    loss_history.append(loss)
    path.append((x,y))

  return loss_history, np.array(path)
```

```python
starting_point = (1.5, 1.5)
epochs = 150
learning_rate = 0.1
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8
loss_history_adam, path_adam = gradient_descent_with_adam(grad_cal, starting_point[0], starting_point[1],
                       beta1, beta2, epochs, learning_rate, epsilon)
```
